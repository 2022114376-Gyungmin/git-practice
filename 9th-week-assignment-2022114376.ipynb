## LAB 8-1:오늘의 날짜와 현재 시간 출력하기

```python
from datetime import datetime

now = datetime.now()

if now.hour < 12:
    am_pm = "오전"
else:
    am_pm = "오후"

hour = now.hour % 12
if hour == 0:
    hour = 12

print(f"오늘의 날짜 : {now.year}년 {now.month}월 {now.day}일")
print(f"현재시간 : {am_pm} {hour}시 {now.minute}분 {now.second}초")
```

## LAB 8-2: 남은 시간 구하기

```python
import datetime as dt

today = dt.date.today()
print(f"오늘은 {today.year}년 {today.month}월 {today.day}일 입니다.")

xMas = dt.datetime(2027, 12, 25)
gap = xMas - dt.datetime.now()

print(f"2027년 크리스마스 까지는 {gap.days}일 {gap.seconds // 3600}시간 남았습니다.")
```

## LAB 8-3: 날짜 계산하기

```python
import datetime as dt

today = dt.datetime.now()

print(f"오늘 날짜 : {today.year}년 {today.month}월 {today.day}일")

thousand = dt.timedelta(days = 1000)
plus1000day = dt.datetime.now() + thousand

print(f"1000일 후 날짜 : {plus1000day.year}년 {plus1000day.month}월 {plus1000day.day}일")
```

```python
import datetime as dt

year, month, day = map(int, input("처음으로 사귄 연도와 월, 일을 입력하시오 : ").split())

first_day = datetime(year, month, day)
couple_day = first_day + dt.timedelta(days = 99)

print(f"100일 기념일은 : {couple_day.year}년 {couple_day.month}월 {couple_day.day}일 입니다.")
```

## LAB 8-5: 랜덤 번호 생성기

```python
import random as rd

numbers = []

while len(numbers) < 3:
    num = rd.randrange(0, 101, 5)

    if num not in numbers:
        numbers.append(num)

numbers.sort()

print("0에서 100 이하의 정수 중에서 5의 배수", numbers)
```

```python
import random as rd

numbers = rd.sample(range(1, 11), 3)

print("1에서 10 사이의 임의의 정수 :", numbers)
```

## LAB 8-6: 로또 번호 얻기

```python
import random as rd

print("이번 주의 추천 로또 번호")

for i in range(5):
    lotto_list = list(range(1, 46))
    rd.shuffle(lotto_list)
    lotto_list = lotto_list[:6]
    lotto_list.sort()
    print(lotto_list)
```

```python
import random as rd

lotto_list = list(range(1, 46))
rd.shuffle(lotto_list)

bonus = lotto_list[6]

lotto_list = lotto_list[:6]
lotto_list.sort()

print("이번 주의 당첨 로또 번호 + 보너스")
print(lotto_list, "+", [bonus])
```

## 연습문제 8.1

```python
from datetime import date

start_day = date(2019, 2, 14)
today = date.today()

passed_day = today - start_day

print(f"춘향이와 몽룡이의 연애 시작일 : {start_day.year}년 {start_day.month}월 {start_day.day}일")
print(f"연애 시작일로부터 경과한 날짜 : {passed_day.days}일")
```

```python
from datetime import date, timedelta

start_day = date(2019, 2, 14)

day100 = start_day + timedelta(days=99)
day200 = start_day + timedelta(days=199)
day500 = start_day + timedelta(days=499)
day1000 = start_day + timedelta(days=999)

print(f"춘향이와 몽룡이의 연애 시작일 : {start_day.year}년 {start_day.month}월 {start_day.day}일")
print(f"100일 기념일 : {day100.year}년 {day100.month}월 {day100.day}일")
print(f"200일 기념일 : {day200.year}년 {day200.month}월 {day200.day}일")
print(f"500일 기념일 : {day500.year}년 {day500.month}월 {day500.day}일")
print(f"1,000일 기념일 : {day1000.year}년 {day1000.month}월 {day1000.day}일")
```

## 연습문제 8.2

```python
import time

def sum1to1000000():
    total = 0

    for i in range(1, 1000001):
        total += i

    return total


start = time.time()

result = sum1to1000000()

end = time.time()

print(f"1에서 1,000,000까지의 합을 구하는 시간 : {end - start:.4f}초")
```

```python
import time

def factorial10000():
    result = 1

    for i in range(1, 10001):
        result *= i

    return result


start = time.time()

for i in range(100):
    factorial10000()

end = time.time()

print(f"10000!을 100번 반복해서 구하는 시간 : {end - start:.4f}초")
```

```python
import time

def odd_cube_sum():
    total = 0

    for i in range(1, 10001, 2):
        total += i ** 3

    return total


start = time.time()

for i in range(100):
    odd_cube_sum()

end = time.time()

print(f"1에서 10000까지 홀수의 세제곱 더하기를 100번 반복하는 시간 : {end - start:.4f}초")
```

```python
import time
import math

def sin_sum():
    total = 0

    for degree in range(1, 361):
        radian = math.radians(degree)
        total += math.sin(radian)

    return total


start = time.time()

for i in range(100):
    sin_sum()

end = time.time()

print(f"1에서 360도까지 sin 값의 합을 100번 반복하는 시간 : {end - start:.4f}초")
```
