```python
# geemap이 잘 작동하는지, 지도를 제대로 불러오는지 확인

import geemap

Map = geemap.Map(center = [35.9156, 128.6014], zoom = 12)
Map
```

```python
# 위성영상 레이어도 추가해보기

# Sentinel-2 영상 1장 가져오기 (대구 지역, 2025년 봄)
image = (ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')     # Sentinel-2 위성영상 전체 불러오기
         .filterBounds(ee.Geometry.Point([128.6014, 35.9156])) # 대구 함지산 좌표값 입력
         .filterDate('2025-04-01', '2025-04-20')               # 4월 1 ~ 20일 영상만 가져오기
         .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 10))  # 구름 10% 미만
         .first()) # 제일 처음 한장만

# 자연색(RGB) 시각화
vis_params = {
    'bands': ['B4', 'B3', 'B2'], # 빨강, 초록, 파랑 = 자연색 / 밴드 설정
    'min': 0,                    # 픽셀 값 명도 설정
    'max': 3000
}

Map = geemap.Map(center=[35.9156, 128.6014], zoom=12)

# 위성영상 가져와서 얹기
# 이 코드가 없으면 위성영상이 안 나옴
Map.addLayer(image, vis_params, 'Sentinel-2 (2025년 4월 초 함지산 산불 직전)')

Map
```
