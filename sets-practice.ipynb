## Exercises: Level 3

```python
age = [22, 19, 24, 25, 26, 24, 25, 24]

age_set = set(age)

print(age)
print(age_set)

print(len(age))
print(len(age_set))

print('set으로 바꾸면 중복되는 숫자는 없어지므로 list인 age가 더 크다.')
```

```python
print('string은 문자열로 글자들이 모여있는 것이며 수정이 불가하다. list는 리스트로 순서가 있고 수정과 중복이 가능하다. tuple은 튜플로 순서가 있고 수정이 불가하나 중복은 가능하다. set는 세트로 순서가 없고 중복이 불가능하다.')
```

```python
sentence = "I am a teacher and I love to inspire and teach people."

words = sentence.split()
unique_words = set(words)

print(words)
print(unique_words)
print(len(unique_words))
```
